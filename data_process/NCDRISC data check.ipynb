```python
import pandas as pd
import glob, io
import zipfile

zip_files = glob.glob('ncdrisc_data/*.zip')

def read_zip(zip_fn):
    zf = zipfile.ZipFile(zip_fn)
    for fn in zf.namelist():
        print(fn)
        if '.csv' in fn:
            return pd.read_csv(zip_fn)
        elif '.xlsx' in fn:
            return pd.read_excel(zf.open(fn).read())

zip_data = []
for zip_file in zip_files:
    df = read_zip(zip_file)
    print(len(df))
    zip_data.append(df)
```

```python
csv_files = glob.glob('ncdrisc_data/*United Kingdom.csv')

result = {}
for csv_file in csv_files:
    print('Check', csv_file)
    try:
        csv_df = pd.read_csv(csv_file, on_bad_lines='skip')
        
        for i, zip_df in enumerate(zip_data):
            try:
                merged = zip_df.merge(csv_df, how='left')
                print('csv df:{}\tzip df:{}\tmerged:{}'.format(len(csv_df), len(zip_df), len(merged)))
                if len(merged) == len(zip_df):
                    result[csv_file] = zip_files[i]
                    break
            except:
                continue
    except:
        print('Fail to load', csv_file)
   
print('Matched:', len(result))
```

```python
for csv_file in csv_files:
    if csv_file not in result:
        print(csv_file)
```
