# Extract the plain text of papers in AHA journals

```python
import requests
from bs4 import BeautifulSoup
import re

from selenium import webdriver 
from selenium.webdriver.chrome.service import Service as ChromeService 
from webdriver_manager.chrome import ChromeDriverManager 

def get_dynamic_page(url):
    driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install())) 
    driver.get(url) 
    return driver.page_source

# Download the html file
url = 'https://www.ahajournals.org/doi/full/10.1161/HYP.0000000000000065?rfr_dat=cr_pub++0pubmed&url_ver=Z39.88-2003&rfr_id=ori%3Arid%3Acrossref.org'
doi = '10.1161/HYP.0000000000000065'
doi = doi.replace('/', '_')
outdir = f'DOI_{doi}'

html_text = get_dynamic_page(url)
html_text
```

```python
# Parse the html file
soup = BeautifulSoup(html_text, 'html.parser')

# Extract the main content of the paper
main_content = soup.find('div', attrs={'class':'hlFld-Fulltext'})

# Extract all paragraphs in the main content
paragraphs = main_content.find_all(['p', 'h3'])
paragraphs
```

```python
# Extract all sections and paragraphs in the main content
section_content = {}
section = ''
body = ''
for p in paragraphs:
    if p.name == 'h3':
        section = p.text
        if section not in section_content:
            section_content[section] = []
    else:
        body = p.text
        section_content[section].append(body)
print(section_content)        
```

```python
# Write sections to files

import os
if not os.path.exists(outdir):
    os.mkdir(outdir)

for section_name, section_body in section_content.items():
    outfile = '{}/{}_section_{}.txt'.format(outdir, doi, section_name.replace('.', '_').replace(' ', '_'))
    with open(outfile, 'w') as fw:
        fw.write('{}\n'.format('\n'.join(section_body)))
```
